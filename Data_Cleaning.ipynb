```python
import os

data_folder_path = '/content/drive/MyDrive/Pollutant Data'
if os.path.exists(data_folder_path):
    print(f"Success! Folder found at: {data_folder_path}")
    print("You can now run Cell 2.")
else:
    print(f"Error: Folder not found at '{data_folder_path}'")
    print("Please double-check the folder name (it's case-sensitive) and path.")
```

```python
input_file = os.path.join(data_folder_path, 'delhi_aqi_timeseries_formatted2.csv')
output_file = os.path.join(data_folder_path, 'delhi_aqi_cleaned.csv')
```

```python
import pandas as pd
import numpy as np
```

```python
try:
  df = pd.read_csv(input_file,parse_dates=['Date'])
except FileNotFoundError:
    print(f"Error: Input file not found at {input_file}")
    print("Please make sure the file exists and the path is correct.")
    raise

```

```python
df.head(20)
```

```python
df.describe()
```

```python
df = df.set_index('Date')
```

```python
df['AQI'] = df['AQI'].interpolate(method='time')
```

```python
df['Pollutants'] = df['Pollutants'].fillna('Unknown')
```

```python
df = df.reset_index()
```

```python
null_count = df['AQI'].isnull().sum()
```

```python
print(f"The number of null values in the 'AQI' column is: {null_count}")
```

```python
null_count_pollutants = df['Pollutants'].isnull().sum()
print(f"the number of null values in the 'Pollutants' column is: {null_count_pollutants}")
```

```python
df.to_csv(output_file, index=False)
```
